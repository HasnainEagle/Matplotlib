# Learning Matplotlib

```python
import numpy as np
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
```

### 2D Plots

* Bivariate Analysis
* categorical -> numerical and numerical -> numerical
* Use case - Time series data

```python
price = [48000,45000,50000,60000,40000,45000]
year = [2015,2016,2017,2018,2019,2020]
plt.plot(year,price)
```

```python
batsman = pd.read_csv("sharma-kohli.csv")
batsman
```

```python
plt.plot(batsman["index"],batsman["RG Sharma"])
```

```python
# plotting multiple plots
plt.plot(batsman["index"],batsman["RG Sharma"])
plt.plot(batsman["index"],batsman["V Kohli"])
```

```python
# labels,Titles
plt.plot(batsman["index"],batsman["RG Sharma"])
plt.plot(batsman["index"],batsman["V Kohli"])
plt.title("Rohit Sharma vs Virat Kohli career")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
# color(hex) and line(style and width) and marker(size)
plt.plot(batsman["index"],batsman["RG Sharma"],color="green",linestyle="dashed")
plt.plot(batsman["index"],batsman["V Kohli"],color="red",linestyle="dotted")
plt.title("Rohit Sharma vs Virat Kohli career")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"],batsman["RG Sharma"],color="green",linestyle="solid",linewidth=4)
plt.plot(batsman["index"],batsman["V Kohli"],color="red",linestyle="solid",linewidth=4)
plt.title("Rohit Sharma vs Virat Kohli career")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"],batsman["RG Sharma"],color="#6A5ACD",linestyle="solid",linewidth=2,marker="o",markersize=10)
plt.plot(batsman["index"],batsman["V Kohli"],color="#FF5733",linestyle="solid",linewidth=2,marker="o")
plt.title("Rohit Sharma vs Virat Kohli career")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
```

```python
plt.plot(batsman["index"],batsman["RG Sharma"],color="#6A5ACD",linestyle="solid",linewidth=2,marker="o",label="Rohit")
plt.plot(batsman["index"],batsman["V Kohli"],color="#FF5733",linestyle="solid",linewidth=2,marker="o",label="Virat")
plt.title("Rohit Sharma vs Virat Kohli career")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend()
```

```python
# Limiting axes
price = [48000,45000,50000,60000,40000,45000,4500000]
year = [2015,2016,2017,2018,2019,2020,2021]
plt.plot(year,price)
plt.ylim(0,75000)
plt.xlim(2015,2019)
# ya hum wab karta han jab hammara dataset ma koi outlier ajaye
```

```python
# grid
plt.plot(batsman["index"],batsman["RG Sharma"],color="#6A5ACD",linestyle="solid",linewidth=2,marker="o",label="Rohit")
plt.plot(batsman["index"],batsman["V Kohli"],color="#FF5733",linestyle="solid",linewidth=2,marker="o",label="Virat")
plt.title("Rohit Sharma vs Virat Kohli career")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend()
plt.grid()
```

```python
# plt.show()
# aghar hum ya graph kisi proper editor pa plot karyengha tu wo plot nahi hogha isliya hum plot.show() use karta han.
plt.plot(batsman["index"],batsman["RG Sharma"],color="#6A5ACD",linestyle="solid",linewidth=2,marker="o",label="Rohit")
plt.plot(batsman["index"],batsman["V Kohli"],color="#FF5733",linestyle="solid",linewidth=2,marker="o",label="Virat")
plt.title("Rohit Sharma vs Virat Kohli career")
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.legend()
plt.grid()
plt.show()
```

### Bar Charts

* Bivariate / Univariate
* numerical vs categorical
* use case - Aggregate Analysis Of Groups

```python
childrens = [14,23,44,56,67]
colors = ["black","blue","orange","red","yellow"]
```

```python
plt.bar(colors,childrens)
```

```python
# horizontal bar chart
plt.barh(colors,childrens)
```

```python
df = pd.read_csv("batsman_season_record.csv")
df
```

```python
# Multiple Bar Charts
```

```python
plt.bar(np.arange(df.shape[0])-0.2,df["2015"],width=0.2,color='blue')
plt.bar(np.arange(df.shape[0]),df["2016"],width=0.2,color="orange")
plt.bar(np.arange(df.shape[0])+0.2,df["2017"],width=0.2,color="green")
# xticks()
plt.xticks(np.arange(df.shape[0]),df["batsman"])
plt.show()
```

```python
childrens = [14,23,44,56,67]
colors = ["blackblackblackblack","blueblueblueblueblueblueblue","orangeorangeorangeorangeorange","redredredredredredred","yellowyellowyellowyellowyello"]
```

```python
plt.bar(colors,childrens)
# overlapping problem
plt.xticks(rotation="vertical")
plt.show()
```

```python
# Stacked Bar Chart
```

```python
plt.bar(df["batsman"],df["2015"],label="2015")
plt.bar(df["batsman"],df["2016"],bottom=df["2015"],label="2016")
plt.bar(df["batsman"],df["2017"],bottom=(df["2015"]+df["2016"]),label="2017")
plt.legend()
```

### Pie Chart

* Univariate/Bivariate
* numerical vs categorical
* use case - to find contribution on a standard scale

```python
data = [100,34,56,78,89]
subjects = ['Maths','ML','Sql','Powerbi','Pandas']
```

```python
plt.pie(data,labels=subjects)
plt.show()
```

```python
df = pd.read_csv("gayle-175.csv")
df
```

```python
plt.pie(df["batsman_runs"],labels = df["batsman"])
plt.show()
```

```python
plt.pie(df["batsman_runs"],labels = df["batsman"],autopct='%0.1f%%')
plt.show()
```

```python
plt.pie(df["batsman_runs"],labels = df["batsman"],autopct='%0.1f%%',colors=["green","red","purple","yellow","cyan","brown"])
plt.show()
```

```python
# explode and shadow
```

```python
plt.pie(df["batsman_runs"],labels = df["batsman"],autopct='%0.1f%%',explode=[0.1,0,0,0,0,0],shadow=True)
plt.show()
```
